```python
from collections import defaultdict, Counter

def build_graph_word(lines, graph=None):
  if not graph:
    graph = defaultdict(Counter)

  for line in lines:
      if line:
        graph['<SOS>'][line[0]] += 1

        for idx in range(0, len(line) - 1):
          curr_token = line[idx]
          next_token = line[idx + 1]

          graph[curr_token][next_token] += 1

        graph[line[-1]]['<EOS>'] += 1

  return graph
```
