```python
l = range(0, 10)
print(l)
print(type(l))

for num in l:
  print(num + 5)

def add_five_func(high_num):
  for i in range(high_num + 1):
    print(i + 5)

add_five_func(9)
```

```python
def count_consecutive_pairs(nums_list):
    counts = {}

    for pair in zip(nums_list, nums_list[1:]):
        counts[pair] = counts.get(pair, 0) + 1

    return counts
```
